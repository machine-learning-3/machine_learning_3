```python
import cv2
import numpy as np
png_img = []
for file in files:
    if '.png' in file: 
        f = cv2.imread(file)
        png_img.append(f)
    if '.PNG' in file: 
        f = cv2.imread(file)
        png_img.append(f)
```

```python
X_vald = np.array(png_img)
```

```python
label = []
for i in range(2000) :
    if i<=199 :
        label.append(0)
    elif i<=399 :
        label.append(1)
    elif i<=599 :
        label.append(2)
    elif i<=799 :
        label.append(3)
    elif i<=999 :
        label.append(4)
    elif i<=1199 :
        label.append(5)
    elif i<=1399 :
        label.append(6)
    elif i<=1599 :
        label.append(7)
    elif i<=1799 :
        label.append(8)
    else :
        label.append(9)
```

```python
from keras.utils import np_utils
Y_vald = np_utils.to_categorical(label)
```

```python
num_classes = 10
(x_train, y_train) , (x_test, y_test) = cifar10.load_data()
y_train = keras.utils.to_categorical(y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)
```

```python
model_d=DenseNet121(weights=None,include_top=False, input_shape=(32, 32, 3))
```

```python
x=model_d.output

x= GlobalAveragePooling2D()(x)
x= BatchNormalization()(x)
x= Dropout(0.5)(x)
x= Dense(1024,activation='relu')(x) 
x= Dense(512,activation='relu')(x) 
x= BatchNormalization()(x)
x= Dropout(0.5)(x)

preds=Dense(10,activation='softmax')(x)
```

```python
model=Model(inputs=model_d.input,outputs=preds)
model.summary()
```

```python
model.compile(optimizer='Adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(x_train/255,y_train,epochs=19,batch_size=32,validation_data=(X_vald/255,Y_vald))
```
